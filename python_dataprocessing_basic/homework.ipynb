```python
# Masukan syntax jawaban di sini
data = pd.DataFrame({
    'no': [1,2,3,4,5],
    'nama_siswa': ['Adi', 'Budi', 'Carli', 'Dadan', 'Endang'],
    'umur': [15.0, 16.0, 14.0, None, 16.0],
    'email': ['andi@gmail.com', 'budi@gmail.com', 'carli@rakamin.com', 'dadan@rakamin.com', 'endang@gmail.com'],
    'score_matematika': [80, 90, 100, 95, 97],
    'score_ekonomi': [90.0, 95.0, 100.0, 86.0, None],
    'score_olahraga': [97, 95, 96, 100, 94],
})

data
```

## No.5

```python
# Masukan syntax jawaban di sini
data_address = pd.DataFrame({
    'nama': ['Adi', 'Budi', 'Dadan', 'Endang', 'Guntur', 'Halim'],
    'alamat': ['DKI Jakarta', 'DKI Jakarta', 'Banten', 'Banten', 'Jawa Barat', 'Aceh']
})

df_merged = new_data.merge(data_address, left_on='nama_siswa', right_on='nama')
# How? inner/outer/left/right?

df_merged.groupby(['alamat']).agg({'umur': 'mean'}).sort_values('umur')
```

```python
# Masukan syntax jawaban di sini
airlines = pd.read_excel('data_airlines.xlsx')
airports = pd.read_excel('data_airports.xlsx')
flights = pd.read_excel('data_flights.xlsx')
```

```python
flights.head()
```

```python
flights.duplicated().any()
```

```python
print(flights['YEAR'].unique()) # Cek apakah ada tahun selain 2015
print(flights['CANCELLED'].unique()) # Cek apakah cancelled itu boolean? 1 0, kalau iya, bisa agregasi pakai sum
```

```python
#a. Total penerbangan per bulan selama 2015 (3 point)
#b. Total penerbangan yang CANCELLED per bulan (3 point)
#c. Rata-rata, Minimum dan Maksimum ELAPSED_TIME per bulan (4 point)
soal_6 = flights.groupby('MONTH').agg({ 'FLIGHT_NUMBER': ['count'],
                                        'CANCELLED': ['sum'],
                                        'ELAPSED_TIME': ['mean', 'min', 'max']
                                         }).reset_index()
soal_6.columns = ['2015_MONTH', 'TOTAL_PENERBANGAN', 'TOTAL_CANCELLED', 'ELAPSED_TIME_MEAN', 'ELAPSED_TIME_MIN', 'ELAPSED_TIME_MAX']
soal_6['SUCCESSFUL_FLIGHT'] = soal_6['TOTAL_PENERBANGAN'] - soal_6['TOTAL_CANCELLED']

soal_6[['2015_MONTH', 'TOTAL_PENERBANGAN', 'TOTAL_CANCELLED', 'SUCCESSFUL_FLIGHT', 'ELAPSED_TIME_MEAN', 'ELAPSED_TIME_MIN', 'ELAPSED_TIME_MAX']]
```

Insight menarik: 
- Winter (bulan desember, januari, februari) banyak penerbangan yang cancelled, cuaca buruk sepertinya.
- Paling sedikit cancel Autumn (September - November)
- Paling banyak penerbangan di tengah Summer (6 **7** 8), karena cuaca sedang panas-panasnya, dan banyak yang liburan ke pantai

```python
# Masukan syntax jawaban di sini
airports[airports['AIRPORT'].str.lower().str.contains('john f. kennedy|northwest florida')]
```

```python
new_flights = flights.merge(airports, how = 'left', left_on='DESTINATION_AIRPORT', right_on='IATA_CODE')
jfk_ecp = new_flights[new_flights['ORIGIN_AIRPORT'].isin(['JFK','ECP'])]
```

```python
soal_7 = jfk_ecp.groupby(['AIRPORT']).agg({'FLIGHT_NUMBER': ['count'],
                                           'CANCELLED': ['sum'],
                                           'DISTANCE': ['mean']
                                          }).reset_index()
soal_7.columns = ['DESTINATION_FROM_JFK_ECP', 'TOTAL_PENERBANGAN', 'TOTAL_CANCELLED', 'DISTANCE_MEAN']
soal_7.sort_values('TOTAL_PENERBANGAN', ascending=False)
```

Destinasi airport mana yang paling banyak jumlah penerbangannya? **San Francisco International Airport**

```python
# Masukan syntax jawaban di sini
tel_revenue = pd.read_csv('01 telecom_revenue.csv')
tel_usage = pd.read_csv('02 telecom_usage.csv')
```

```python
print('Data revenue ada duplikat?', tel_revenue.duplicated(subset='CustomerID').any())
tel_revenue.head()
```

```python
print('Data usage ada duplikat?', tel_usage.duplicated(subset='CustomerID').any())
tel_usage.head()
```

```python
monthly_revenue = tel_revenue['MonthlyRevenue'] > 10
occupation = tel_revenue['Occupation'].isin(['Professional', 'Student', 'Crafts'])
df1 = tel_revenue[monthly_revenue & occupation]

awal_genap = tel_usage['CustomerCareCalls'].astype(str).str.startswith(('2', '4', '6', '8'))
un_more_block = tel_usage['UnansweredCalls'] > tel_usage['BlockedCalls']
df2 = tel_usage[awal_genap & un_more_block]
```

```python
soal_8 = df1.merge(df2, on='CustomerID')
```

```python
soal_8_agg = soal_8.groupby('Occupation').agg({'CustomerID': 'nunique',
                                               'DroppedCalls': 'mean'
                                              }).reset_index()
soal_8_agg.columns = ['Occupation', 'total_CustomerID', 'mean_DroppedCalls']
soal_8_agg
```

pada occupation apa, rata-rata DroppedCalls paling besar? **Crafts**
