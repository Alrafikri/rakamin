# Soal 4

Buatlah sebuah function string_processor dengan inputan 2 parameter, yaitu suatu string 1 kata dan nama proses yang akan dilakukan. Function ini memiliki 4 proses

```python
def string_processor(kata, proses='first_char'):
    if proses == 'first_char':
        return kata[0]
    elif proses == 'last_char':
        return kata[-1]
    elif proses == 'num_chars':
        return len(''.join(kata.split()))
    elif proses == 'num_vowels':
        kata_vowel = [huruf for huruf in kata if huruf.lower() in 'aiueo']
        return len(kata_vowel)
    else:
        print('input proses salah')
```

```python
string_processor('''TugAs ProgrAm
                    mIng Pyth         On''','first_char')
```

```python
string_processor('''TugAs ProgrAm
                    mIng Pyth         On''','last_char')
```

```python
# Ini function hitung salesnya, dengan parameter default 2017
def hitung_sales(region, tahun=2017, df=df_superstore):
    df['order_year'] = df['Order Date'].dt.year
    filter_region = df['Region'] == region
    filter_tahun = df['order_year'] == tahun
    return df[filter_region & filter_tahun].groupby('order_year').agg({'Sales': 'sum',
                                                                       'Profit': 'sum'
                                                                      }).reset_index()
```

```python
# data dikelompokkan berdasarkan customer ID -> karena ada 1 customer berulang kali order
# Hati-hati, karena ini segmentasi user maka perlu melakukan agregasi di level user terlebih dulu sebelum melakukan segmentasi
df_customer = df_superstore[['Customer ID','Sales', 'Quantity', 'Discount', 'Profit']].groupby('Customer ID').agg({'Sales': 'sum',
                                                                                                                   'Quantity': 'sum',
                                                                                                                   'Discount': 'mean',
                                                                                                                   'Profit': 'sum'
                                                                                                                  }).reset_index()

df_customer.head()
```

```python
df_customer.describe().T
```

*Buatlah 3 kelompok 'segment' (high, medium, dan low) berdasarkan performance Sales*

3 kelompok segmen dibuat berdasarkan quartile. Alasannya supaya pembagian tiap kelompok dapat terukur berapa kira-kira yang masuk kelompok tersebut.

* Q1 ke bawah = 25% pengguna (Low)
* Q1-Q3 = 50% pengguna (Medium/average)
* Q4 ke atas = 25% pengguna (High)

Apabila dilihat dari sales, maka:
* Low = sales kurang dari 1146.05
* Medium = sales di antara 1146.05 - 3785.276
* High = sales lebih dari 3785.276

```python
# ini fungsi segmentasinya
def segmentasi(df):
    Q1 = df_customer['Sales'].quantile(q = 0.25)
    Q3 = df_customer['Sales'].quantile(q = 0.75)
    try:
        if df['Sales'] < Q1:
            return 'Low'
        elif df['Sales'] > Q3:
            return 'High'
        else:
            return 'Medium'
    except:
        return 'Other'
```

```python
# Tentukan jumlah customer pada masing-masing segment di tahun 2017!

# fungsi di bawah ini digunakan untuk mengelompokkan pembelian user per-tahun (groupby order_year, customer ID), jadi bisa difilter total pembelian user dalam tahun tertentu.
def kelompokkan_tahun(df=df_superstore):
    df['order_year'] = df['Order Date'].dt.year
    return df[['order_year','Customer ID','Sales', 'Quantity', 'Discount', 'Profit']].groupby(['order_year','Customer ID']).agg({'Sales': 'sum',
                                                                                                                                 'Quantity': 'sum',
                                                                                                                                 'Discount': 'mean',
                                                                                                                                 'Profit': 'sum'
                                                                                                                                }).reset_index()

df_cust_year = kelompokkan_tahun()
df_cust_year.head()
```

```python
# menggunakan apply lambda untuk membuat baris segmentasi
df_cust_year['segmentasi'] = df_cust_year.apply(lambda x: segmentasi(x), axis = 1)
```

```python
# anyway, profit tidak sebanding dengan sales. Tapi karena segmentasi disuruh berdasarkan sales, jadi hasilnya ada yang profit tinggi tapi Low, profit rendah tapi medium.
df_cust_year.head()
```

```python
# di groupby berdasarkan segmentasinya untuk mengetahui jumlah user (nunique customer ID) dalam tahun 2017 (filter)
df_cust_year[df_cust_year['order_year'] == 2017].groupby('segmentasi').agg({'Customer ID': 'nunique'}).reset_index()
```

```python
import pandas as pd
import numpy as np
from datetime import date

def rfm_analysis(df,r,f,m,method='simple', weight=[1/3,1/3,1/3],grid='fr'):
    """
    Objective:
    Consumer segmentation based on a simple RFM analysis
    
    Arguments:
    - df = dataframe to segment, type pandas.DataFrame
    - r = column name for recency data, type string
    - f = column name for frequency data, type string
    - m = column name for monetary data, type string
    - method = {'simple' or 'fr-grid'}, default 'simple'
    - weight = list containing weight of each rfm variables, used if method = 'simple'
        expected value = [r_weight, f_weight, m_weight], default [1/3,1/3,1/3]
    
    Return: Pandas.DataFrame containing segmentation label
    """
    
    try:
        df = df.copy()
        df['r_rank'] = 5 - pd.qcut(df[r], q=5, labels = False)
        df['f_rank'] = pd.qcut(df[f], q=5, labels = False) + 1
        df['m_rank'] = pd.qcut(df[m], q=5, labels = False) + 1
        if(method == 'simple'):
            if(sum(weight) != 1):
                raise ValueError('Jumlah variabel weight harus 1')
            else:
                df['rfm_score'] = np.average(df[['r_rank','f_rank', 'm_rank']], weights=weight, axis=1)
                bins = [0, 1.6, 3, 4, 4.5, 5]
                names = ['Lost Customer', 'Low-value Customer', 'Medium-value Customer', 'High-value Customer', 'Top Customer']
                df['segment'] = pd.cut(df['rfm_score'], bins, labels=names)
                return df
        elif(method == 'fr-grid'):
            fr_grid = [
                ['Hibernating', 'Hibernating', 'About To Sleep', 'Promising', 'New customers'],
                ['Hibernating', 'Hibernating', 'About To Sleep', 'Potential Loyalist', 'Potential Loyalist'],
                ['At Risk', 'At Risk', 'Need Attention', 'Potential Loyalist', 'Potential Loyalist'],
                ['At Risk', 'At Risk', 'Loyal Customers', 'Loyal Customers', 'Champions'],
                ["Can't Lose Them", "Can't Lose Them", 'Loyal Customers', 'Loyal Customers', 'Champions']
            ] # sometimes the dumbest method win.
            df['segment'] = df[['f_rank', 'r_rank']].apply(lambda x: fr_grid[x.f_rank - 1][x.r_rank - 1], axis = 1)
            return df
    except:
        print('Something went wrong')
```
